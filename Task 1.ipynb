### Create a random 100000x1 array from the same exponential distribution and visualize the distribution histogram along with uniform and normal distributions. Use the code given below and do the necessary changes. Change the number of bins and the plot view to make the visualization better.

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
plt.hist(np.random.exponential(scale=1.0,size=100000), density=True, bins=100, histtype="step", color="blue", label="Exponential")
plt.axis([-2.5, 2.5, 0, 1.1])
plt.legend(loc = "upper left")
plt.title("Random distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

### Using matplotlib, plot the Z = X ** 2 + Y ** 2 in a 3D plot. Limit the x and y values to [-5, 5]. 

```python
from mpl_toolkits import mplot3d
```

```python
def my_function_Z(x,y):
    return x**2+y**2

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

X, Y = np.meshgrid(x, y)
Z = my_function_Z(X, Y)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
```
